```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
from sklearn import preprocessing
df1 = pd.DataFrame(np.random.randint(0,1000,size=(100, 1)), columns=list('A'))
df2 = pd.DataFrame(np.random.randint(0,10,size=(100, 1)), columns=list('B'))
#df3=df2/100
df4=pd.concat([df1,df2],axis=1)
import matplotlib.pyplot as plt
df4['label']=np.where((df4['A']>=500) & (df4['B']>=5),'yes','no')
print(df4['label'].value_counts())
sns.FacetGrid(df4,hue='label',size=6).map(plt.scatter,'A','B').add_legend() 
plt.show()
y=df4['label'] 
X=df4.drop(columns=['label'])
#print(X.head()) 
#print(y.head())
scaler = preprocessing.Normalizer()
scaled_df = scaler.fit_transform(X)
df10=pd.DataFrame(scaled_df)
#print(df10.head())
newdf=pd.concat([df10,y],axis=1)
#print(newdf.head())
sns.FacetGrid(newdf,hue='label',size=6).map(plt.scatter,0,1).add_legend()
plt.show()
```
